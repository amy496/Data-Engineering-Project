```python
import numpy as np
import pandas as pd
from bokeh.io import show, output_file
from bokeh.plotting import figure,show
import pandas_bokeh
from bokeh.models import ColumnDataSource,HoverTool
from bokeh.palettes import Spectral6
from bokeh.transform import factor_cmap

from bokeh.plotting import figure

import requests

#_url = "http://lookup-service-prod.mlb.com/json/named.team_all_season.bam?sport_code='mlb'&season=2021"
_url =   "http://lookup-service-prod.mlb.com/json/named.roster_40.bam?team_id=147"

       
resp = requests.get(_url)
#if resp.status_code == 200:
    #print(resp.text)
x = resp.json()
# df = pd.DataFrame(x["named.team_all_season.bam?sport_code='mlb'&season=2021"])
# df  
#create a bokeh chart
#integrate
Yankees_roster_2017 = x["roster_40"]['queryResults']['row']   
   
```

```python
df = pd.DataFrame(Yankees_roster_2017)
df
```

```python
stats = []
for id in df['player_id']:
    _url = f"http://lookup-service-prod.mlb.com/json/named.sport_hitting_tm.bam?league_list_id='mlb'&game_type='R'&season='2021'&player_id='{id}'"
    resp = requests.get(_url)
    
    x = resp.json()
    stats.append(x['sport_hitting_tm']['queryResults'].get('row'))
```

```python
player_stats = (list(filter(lambda x: x != None,stats)))
```

```python
for idx, s in enumerate(player_stats):
     if type(s) != dict:
        player_stats[idx] = s[1]
```

```python
baseball_player_stats = pd.DataFrame(player_stats)
```

```python
baseball_player_stats
```

```python
final_player_stats = pd.merge(df,baseball_player_stats, on='player_id')
```

```python
final_player_stats
```

```python
final_player_stats.drop(columns=['college','starter_sw', 'end_date_x','name_first'], axis=1)
```

```python


grouped = final_player_stats.groupby('band_name')['fans'].sum().sort_values(ascending=False).to_frame().iloc[:10]
grouped
```

```python
final_player_stats[final_player_stats['position_txt'] != "P"]
```

```python
yankee_stats = final_player_stats[final_player_stats['position_txt'] != "P"]
yankee_stats
```

```python
grouped = yankee_stats.groupby('name_display_first_last')['obp'].sum().sort_values(ascending=True).to_frame()
grouped
```

```python
colors=["#FF00FF","#F08080", '#FFFFE0', "#00FF00","#B0E0E6", "#FF69B4"," #4682B4","#FFFAFA","#DEB887","#A0522D","#FFDAB9","#FF6347","#800080","#FFA07A","#FFD700"]
```

```python
grouped['colors'] = colors
```

```python
source = ColumnDataSource(data=grouped)
source
```

```python
player_values = source.data['name_display_first_last'].tolist()
print(player_values)
p = figure(y_range=player_values, plot_width=1000,plot_height=700)
```

```python
p.hbar(y='name_display_first_last', right='obp', source=source, width=0.8, color='colors')
p.add_tools(HoverTool())

p.title.text ="Yankees On Base Percentage 2021"
p.xaxis.axis_label = 'On Base Percentage'
p.yaxis.axis_label = 'Yankee Players'

show(p)
```
